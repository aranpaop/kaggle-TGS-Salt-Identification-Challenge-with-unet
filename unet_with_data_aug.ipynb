```python
import cv2
import numpy as np
import os
from imp import reload
import pandas as pd

from keras.models import load_model

from snn import unet
from snn import processor
from snn import batch_generator
```

```python
reload(unet)
reload(processor)
reload(batch_generator)
```

```python
X_train_path = 'train/images'
y_train_path = 'train/masks'
```

```python
X_train = []
y_train = []
for filename in os.listdir(X_train_path):
    img = cv2.imread(os.path.join(X_train_path, filename), 0)
    img = cv2.resize(img, (128, 128))
    img = img / 255
    X_train.append(img)
    mask = cv2.imread(os.path.join(y_train_path, filename), 0)
    mask = cv2.resize(mask, (128, 128), interpolation=cv2.INTER_NEAREST)
    mask = mask / 255
    y_train.append(mask)
```

```python
model = unet.get_unet()
```

```python
epochs = 40000
batch_size = 20
best_iou = 0
iou_list = []
loss_list = []
tol = 0
for epoch in range(epochs):
    X, y = batch_generator.batch_gen(X_train, y_train, batch_size)
    [loss, iou] = model.train_on_batch(X, y)
    loss_list.append(loss)
    iou_list.append(iou)
    if epoch % 50 == 49:
        iou = sum(iou_list) / 50
        if iou > best_iou:
            best_iou = iou
            cur_loss = sum(loss_list) / 50
            model.save('aug_unet_model.h5')
            print('saving model')
            tol = 0
        else:
            tol += 1
        if tol == 20:
            break
        loss_list = []
        iou_list = []
        print('Epoch '+str(epoch+1)+' / '+str(epochs))
        print('best iou: '+str(best_iou)+' with loss: '+str(cur_loss))
print('training complete, with best iou: '+str(best_iou)+', and loss: '+str(cur_loss))
```

```python
model = load_model('aug_unet_model.h5', custom_objects={'mean_iou': processor.mean_iou})
```

```python
X = np.array(X_train)[:, :, :, np.newaxis]
y = np.array(y_train)[:, :, :, np.newaxis]
```

```python
y_pred = model.predict(X, verbose=1)
```

```python
score_v_thre = []
for threshold in np.arange(0.5, 1.0, 0.05):
    score_v_thre.append(processor.calc_score(y, y_pred, threshold))
```

```python
best_threshold = 0.5 + score_v_thre.index(max(score_v_thre)) * 0.05
```

```python
X_test_path = 'test/images'
X_test = []
for filename in os.listdir(X_test_path):
    img = cv2.imread(os.path.join(X_test_path, filename), 0)
    img = cv2.resize(img, (128, 128))
    img = img / 255
    X_test.append(img)
X_test = np.array(X_test)[:, :, :, np.newaxis]
```

```python
y_test_pred = model.predict(X_test, verbose=1)
```

```python
img_cnt = 0
pre_list = []
for filename in os.listdir(X_test_path):
    if img_cnt % 100 == 0:
        print(str(img_cnt)+' data listed')
    img = y_test_pred[img_cnt, :, :, :]
    img.reshape((128, 128))
    img = cv2.resize(img, (101, 101))
    pre_list.append((filename[:-4], processor.pic2step(img, best_threshold)))
    img_cnt += 1
```

```python
df = pd.DataFrame(data=pre_list, columns=['id', 'rle_mask'])
df.to_csv('submission4.csv', index=False, header=True)
```
